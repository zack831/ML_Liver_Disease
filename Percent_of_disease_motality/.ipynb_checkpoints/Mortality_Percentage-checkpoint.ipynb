```python
died_cr = len(df.loc[(df['CLASS'] == 1) & (df['STATUS'] == 1)])
died_hp = len(df.loc[(df['CLASS'] == 2) & (df['STATUS'] == 1)])
died_hc = len(df.loc[(df['CLASS'] == 3) & (df['STATUS'] == 1)])
total = len(df)

percent = []
data = [died_cr, died_hp, died_hc]

for i in data:
    percent.append(((i/total)*100))

print('For Cirrhosis Mortality percentage:',percent[0])
print('For Hepatitis Mortality percentage::',percent[1])
print('For Hcc Mortality percentage::',percent[2])
```
