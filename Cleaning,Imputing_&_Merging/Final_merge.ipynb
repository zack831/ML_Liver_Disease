```python
csv_files = [('/home/zack83/Dropbox/Tripura_University/Project_Final/Databases/Merged_Part2/cirrFull.csv'),('/home/zack83/Dropbox/Tripura_University/Project_Final/Databases/Merged_Part2/hepaFull.csv'),('/home/zack83/Dropbox/Tripura_University/Project_Final/Databases/Cleaned_Part2/hccF.csv')]
full = pd.concat([pd.read_csv(f) for f in csv_files ], ignore_index=True)
```

```python
full.info()
```

```python
full.select_dtypes(include=(['int64', 'float64'])).isna().sum()
df_num_col = full.select_dtypes(include=(['int64', 'float64'])).columns
for c in df_num_col:
    full[c].fillna(full[c].median(), inplace=True)

full.select_dtypes(include=(['int64', 'float64'])).isna().sum()
```

```python
full.info()
```
